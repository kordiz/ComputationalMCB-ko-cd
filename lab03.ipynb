```python
import sys
import string
from matplotlib import pyplot as plt
import numpy as np
import operator
```

```python
word_list = []
with open('Downloads/romeoandjuliet.txt', 'r') as rj:
    word_list = [word for line in rj for word in line.split()]
    rj.close()

word_dict = {}
for word in word_list:
    word = word.lower()
    punc = set(string.punctuation)
    word = ''.join(ch for ch in word if ch not in punc)
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1

    
```

```python
keys = [key for key in word_dict.keys()]
values = [int(word_dict[key]) for key in keys]


plt.figure(figsize = (20,5))
plt.bar(keys, values)
plt.xticks(keys, keys, rotation='vertical')
plt.xlabel('word')
plt.ylabel('word count')
plt.title('word vs word count')
plt.show()
```

```python
sorted_keys = sorted(word_dict.items(), key=operator.itemgetter(1))
most = [item[0] for item in sorted_keys[-10: -1][::-1]]
least = [item[0] for item in sorted_keys[0:10]]


print("most used words: " + str(most))
print('least used words: ' + str(least))
```

The most used words are articles/common verbs, which is not surprising. The least used words printed above are not a complete list of the least used words, since more than 10 words were used only once. These are the 10 words that were only used once that came first in the .txt file. However, these words are all fairly uncommon, or may be more commonly used in a different form (eg plural).

```python
sorted_keys = sorted(word_dict.items(), key=operator.itemgetter(1), reverse=True)
sort_keys = [item[0] for item in sorted_keys]
sort_values = [int(item[1]) for item in sorted_keys]

plt.plot(np.log(sort_values), np.log(range(len(sort_values))))
plt.xlabel('log (word count)')
plt.ylabel('log (word rank)')
plt.title('log (word count) vs log (word rank)')
plt.show()
```

This plot shows a nearly linear negative relationship between the log (word count) and log (word rank), which means that rank and word count both decrease exponentially. The most popular words are very frequently used, and the least popular words are used very infrequently.
